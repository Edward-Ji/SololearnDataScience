# DataVisualisation

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

The figure contains all the objects, including axes, graphics, texts, and labels. The axes is a bounding box with ticks and labels. Think of axes as an individual plot.

```python
fig = plt.figure()
ax = plt.axes()
plt.show()
```

prepare data

```python
x = np.linspace(0, 10, 1000)
y = np.sin(x)
```

### plot using `axes` or `plt` directly

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.show()
```

### labels and titles

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('graph of y=sin(x)')
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim((0, 2*np.pi))
plt.ylim((-1.5, 1.5))
plt.title('graph of y=sin(x)')
plt.show()
```

### multiple lines

```python
x = np.linspace(0,10,1000) # 1darray of length 1000
plt.plot(x, np.sin(x), color='k')
plt.plot(x, np.cos(x), color='r', linestyle ='--')
plt.show()
```

Refer to [Documentation](https://matplotlib.org/2.0.2/api/colors_api.html) for all colors
***
It is usually useful to create a legend for multiple lines.

```python
x = np.linspace(0,10,1000) # 1darray of length 1000
plt.plot(x, np.sin(x), 'k:', label='sin(x)')
plt.plot(x, np.cos(x), 'r--', label='cos(x)')
plt.legend()
plt.show()
```

## Plotting

```python
presidents_df = pd.read_csv('https://sololearn.com/uploads/files/president_heights_party.csv',
                            index_col='name')
```

above code prepare data introduced DataAnalysis page

```python
plt.scatter(presidents_df['height'], presidents_df['age'])
plt.show()
```

```python
plt.scatter(presidents_df['height'], presidents_df['age'],
            marker='<',
            color='k')
plt.xlabel('height'); 
plt.ylabel('age')
plt.title('U.S. presidents')
plt.show()
```

Check [Documentation](https://matplotlib.org/api/markers_api.html?highlight=marker#module-matplotlib.markers) for all markers
***
Pandas integrates with matplotlib, so we can plot directly from a pandas DataFrame object.

```python
presidents_df.plot(kind = 'scatter', 
                   x = 'height', 
                   y = 'age',
                   title = 'U.S. presidents')
plt.show()
```

```python
presidents_df['height'].plot(kind='hist',
                             title = 'height',
                             bins=5)

presidents_df.plot(kind='hist',
                   y='height',
                   title = 'height',
                   bins=5)


plt.show()
```

```python
plt.hist(presidents_df['height'], bins=5)
```

`plt.hist` also returns 1darrays of frequency and bin end points
\
A histogram shows the underlying frequency distribution of a set of continuous data, allowing for inspection of the data for its shape, outliers, skewness, etc.

```python
presidents_df['height'].describe()
```

```python
plt.style.use('classic')
presidents_df.plot(kind='box',
                   y='height')
plt.style.use('default')
```

In classic style, red bar represents the median.

```python
party_cnt = presidents_df['party'].value_counts()

party_cnt.plot(kind='bar')
```

Bar plots are different from histogram. Histogram presents numeric data, whereas bar plot shows categorical data. As a result, histogram has no gap between bars, unlike that in bar plot.
